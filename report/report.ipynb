```python
import random


def generate_dataset(num_nodes, time_limit_min, time_limit_max):
    dataset = []
    for node in range(1, num_nodes + 1):
        if random.randint(1, 10) > 8:
            time = random.randint(0, time_limit_max) + random.random()
        else:
            time = random.randint(time_limit_min, time_limit_max) + random.random()
        
        dataset.append({'node': node, 'time': time})
    return dataset



num_nodes = 1024*5  #  
time_limit = 100  #  
time_dataset = generate_dataset(num_nodes, 25, 34)
for data in time_dataset:
    print(data)
```

```python
import matplotlib.pyplot as plt

def plot_dataset_grouped(dataset):
    nodes = [data['node'] for data in dataset]
    times = [data['time'] for data in dataset]

    plt.barh(range(len(nodes)), times, color='skyblue')
    plt.yticks(['' for _ in nodes])
    plt.xlabel('Time')
    plt.title('Time of BOD')
    plt.show()

#   
plot_dataset_grouped(time_dataset)
```

```python
def plot_dataset_boxplot(dataset):
    times = [data['time'] for data in dataset]

    plt.boxplot(times)
    plt.ylabel('Time (sec)')
    plt.title('BOD execution time')
    plt.show()

#   
plot_dataset_boxplot(time_dataset)
```

```python
def generate_dataset_solutions(num_nodes, min_solutions, max_solutions):
    dataset = []
    for node in range(1, num_nodes + 1):
        if node > 100 and node < 1000:
            if random.randint(1, 1000) > 950:
                num_solutions = random.randint(0, max_solutions+3)
                dataset.append({'node': node, 'solutions': num_solutions})
            if random.randint(1, 1000) > 400:
                num_solutions = random.randint(-2, max_solutions)
                dataset.append({'node': node, 'solutions': num_solutions})
            continue
        if node > 3000 and node < 4000:
            if random.randint(1, 1000) > 950:
                num_solutions = random.randint(0, max_solutions+3)
                dataset.append({'node': node, 'solutions': num_solutions})
            if random.randint(1, 1000) > 400:
                num_solutions = random.randint(-2, max_solutions)
                dataset.append({'node': node, 'solutions': num_solutions})
            continue
        if random.randint(1, 1000) > 950:
            num_solutions = random.randint(0, max_solutions+3)
        elif random.randint(1, 1000) > 950:
            num_solutions = random.randint(-6, max_solutions+3)
        else:
            num_solutions = random.randint(min_solutions, max_solutions)
        dataset.append({'node': node, 'solutions': num_solutions})
    return dataset
```

```python
sol_df = generate_dataset_solutions(1024*5, -4, 5)

def plot_dataset_solutions(dataset):
    nodes = [data['node'] for data in dataset]
    solutions = [data['solutions'] for data in dataset]

    colors = ['green' if s >= 0 else 'red' for s in solutions]

    plt.bar(nodes, solutions, color=colors)
    plt.xlabel('Node') 
    plt.ylabel('Number of Solutions')
    plt.title('Difference in solutions found')
    plt.show()


plot_dataset_solutions(sol_df)
```

```python
def generate_dataset_good_bad_solutions(num_nodes, max_good_solutions, max_bad_solutions):
    dataset = []
    for node in range(1, num_nodes + 1):
        good_solutions = random.randint(0, max_good_solutions)
        bad_solutions = random.randint(0, max_bad_solutions)
        dataset.append({'node': node, 'good_solutions': good_solutions, 'bad_solutions': bad_solutions})
    return dataset


df = generate_dataset_good_bad_solutions(100, 4, 3)
```

```python
import numpy as np


def plot_dataset_good_bad_solutions(dataset):
    nodes = [data['node'] for data in dataset]
    good_solutions = [data['good_solutions'] for data in dataset]
    bad_solutions = [data['bad_solutions'] for data in dataset]

    #  
    bar_width = 0.5

    #   
    index = np.arange(len(nodes))

    plt.bar(index, bad_solutions, bar_width, color='red', label='Bad Solutions')
    plt.bar(index, good_solutions, bar_width, color='green', label='Good Solutions', bottom=bad_solutions)

    plt.xlabel('Node')
    plt.ylabel('Number of Solutions')
    plt.title('Good and Bad Solutions')
    plt.legend()
    plt.show()

plot_dataset_good_bad_solutions(df)
```

```python
def plot_dataset_difference(dataset):
    nodes = [data['node'] for data in dataset]
    good_solutions = [data['good_solutions'] for data in dataset]
    bad_solutions = [data['bad_solutions'] for data in dataset]

    differences = [good - bad for good, bad in zip(good_solutions, bad_solutions)]
    colors = ['green' if diff >= 0 else 'red' for diff in differences]

    plt.bar(nodes, differences, color=colors)
    plt.xlabel('Node')
    plt.ylabel('Difference (Good - Bad Solutions)')
    plt.title('Difference between Good and Bad Solutions')
    plt.show()

plot_dataset_difference(df)
```
